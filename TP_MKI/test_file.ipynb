```python
video_repo = pathways.videos()
mm_cad1 = os.path.join(video_repo, "mm_cad1_live_output")

pairs = []

for action in os.listdir(mm_cad1):
    action_path = os.path.join(mm_cad1, action)
    if not os.path.isdir(action_path):
        continue

    videos, jsons = {}, {}

    for f in os.listdir(action_path):
        full_path = os.path.normpath(os.path.join(action_path, f))
        if f.endswith(".mp4"):
            key = os.path.splitext(f)[0]
            videos[key] = full_path
        elif f.endswith(".json"):
            key = os.path.splitext(f)[0]
            jsons[key] = full_path

    for key, vpath in videos.items():
        if key in jsons:
            pairs.append({
                "action": action,
                "video": vpath,
                "json": jsons[key]
            })
```

```python
frame_nb = []
short_videos = []
medium_frames = []
long_videos = []
for i in range(len(pairs)):
    json_path = pairs[i]['json']

    with open(json_path, "r") as f:
        data = json.load(f)
    
    frame_nb.append(len(data['frames']))  # Print the number of frames in each JSON file

    if len(data['frames']) < 10:
        short_videos.append([pairs[i]['video'], len(data['frames'])])
    
    if 10 <= len(data['frames']) <= 16:
        medium_frames.append([pairs[i]['video'], len(data['frames'])])

    if 30 <= len(data['frames']):
        long_videos.append([pairs[i]['video'], len(data['frames'])])

    
```

```python
import statistics

min_frames = min(frame_nb)
max_frames = max(frame_nb)
mean_frames = statistics.mean(frame_nb)
median_frames = statistics.median(frame_nb)
stdev_frames = statistics.stdev(frame_nb)

print(f"Min frames: {min_frames}")
print(f"Max frames: {max_frames}")
print(f"Mean frames: {mean_frames}")
print(f"Median frames: {median_frames}")
print(f"Standard Deviation of frames: {stdev_frames}")
```
